# Importing original dataset

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading csv file
df = pd.read_csv("dataset/eurusd_minute.csv")
print(df.head(10))
```

# Calculating Signals

```python
#creating empty array for signals
signals=[np.nan for x in range(len(df['Close']))]
```

```python
#calculating signals
# 1 if prev close is less than current
# -1 if prev close is greater than current
# 0 if prev close is equal to current
for i in range (1,len(df['Close'])-1):
    if df['Close'][i-1]>df['Close'][i]:
        signals[i]=-1
    elif df['Close'][i-1]<df['Close'][i]:
        signals[i]=1
    else:
        signals[i]=0
```

```python
#saving calculate signals to dataframe
df['signal']=signals
```

```python
#removing first and last value from dataset
df = df.drop([df.index[0] , df.index[-1]]).reset_index(drop=True)
#print(df[['Close','signal']].head(50))
#print(df[['Close','signal']].tail(50))
```

```python
df.to_csv('processedData/eur_usd_1min_1_close.csv',index=False)
```

# Dataset with prev 3 close

```python
#creating dataframe
df2=pd.DataFrame()
#print(df2)
```

```python
def calculatePrev3(data):
    prev_1=[np.nan for x in range(len(data))]
    prev_2=[np.nan for x in range(len(data))]
    prev_3=[np.nan for x in range(len(data))]
    for i in range(3,len(data)):
        prev_1[i]=data[i-1]
        prev_2[i]=data[i-2]
        prev_3[i]=data[i-3]
    return prev_1,prev_2,prev_3    
```

```python
#creating prev 3 closes

prev_close_1,prev_close_2,prev_close_3 = calculatePrev3(df['Close'])
```

```python
df2['Date']=df['Date']
df2['Open']=df['Open']
df2['High']=df['High']
df2['Low']=df['Low']
df2['Close']=df['Close']
df2['prev_close_1']=prev_close_1
df2['prev_close_2']=prev_close_2
df2['prev_close_3']=prev_close_3
df2['signal']=df['signal']
```

# Calculating Indicators 

```python
#RSI funtion
def computeRSI (data, window):
    diff = data.diff(1).dropna()# diff in one field(one day)

    #this preservers dimensions off diff values
    up_chg = 0 * diff
    down_chg = 0 * diff
    
    # up change is equal to the positive difference, otherwise equal to zero
    up_chg[diff > 0] = diff[ diff>0 ]
    
    # down change is equal to negative deifference, otherwise equal to zero
    down_chg[diff < 0] = diff[ diff < 0 ]
    
    up_chg_avg   = up_chg.ewm(com=window-1 , min_periods=window).mean()
    down_chg_avg = down_chg.ewm(com=window-1 , min_periods=window).mean()
    
    rs = abs(up_chg_avg/down_chg_avg)
    rsi = 100 - 100/(1+rs)
    return rsi

#calculating RSI based on close values
rsi_14=computeRSI(df['Close'],14)
rsi_20=computeRSI(df['Close'],20)
```

```python
#bollinger band function
def computeBOLINGER(data, window):
    mov_avg = data.rolling(window=window).mean()
    stn_dev = data.rolling(window=window).std() 

    upper = mov_avg + (stn_dev * 2)
    lower = mov_avg - (stn_dev * 2)
    
    return mov_avg,upper,lower

#calculating Bollinger band values based on close values
boll_20, boll_20_ub, boll_20_lb=computeBOLINGER(df['Close'],20)
boll_14, boll_14_ub, boll_14_lb=computeBOLINGER(df['Close'],14)
```

```python
#calculating EMA based on close values
ema_10 = df['Close'].ewm(span=10, adjust=False).mean()
ema_25 = df['Close'].ewm(span=25, adjust=False).mean()
ema_50 = df['Close'].ewm(span=50, adjust=False).mean()
ema_100 = df['Close'].ewm(span=100, adjust=False).mean()
ema_200 = df['Close'].ewm(span=200, adjust=False).mean()
```
